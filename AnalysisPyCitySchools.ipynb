## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
ninth_grade = student_data.loc[student_data["grade"] == "9th",["student_name", "school_name","math_score"]]
tenth_grade = student_data.loc[student_data["grade"] == "10th",["student_name", "school_name","math_score"]]
eleventh_grade = student_data.loc[student_data["grade"] == "11th",["student_name", "school_name","math_score"]]
twelth_grade = student_data.loc[student_data["grade"] == "12th",["student_name", "school_name", "math_score"]]
twelth_grade.head()
```

```python
ninth_group = ninth_grade.groupby(["school_name"])
ninth_summary = ninth_group.mean()
tenth_group = tenth_grade.groupby(["school_name"])
tenth_summary = tenth_group.mean()
eleventh_group = eleventh_grade.groupby(["school_name"])
eleventh_summary = eleventh_group.mean()
twelth_group = twelth_grade.groupby(["school_name"])
twelth_summary = twelth_group.mean()
twelth_summary.head()
```

```python
ninth_and_tenth = pd.merge(ninth_summary, tenth_summary, how="outer", on=["school_name"])
ninth_and_tenth = ninth_and_tenth.rename(columns={"math_score_x": "9th math scores",
                                                 "math_score_y": "10th math scores"})
ninth_and_tenth.head()
```

```python
eleventh_and_twelth = pd.merge(eleventh_summary, twelth_summary, how="outer", on=["school_name"])
eleventh_and_twelth = eleventh_and_twelth.rename(columns={"math_score_x": "11th math scores",
                                                 "math_score_y": "12th math scores"})
eleventh_and_twelth.head()
```

```python
all_math_grades = pd.merge(ninth_and_tenth, eleventh_and_twelth, how="outer", on=["school_name"])
all_math_grades.head()
```
